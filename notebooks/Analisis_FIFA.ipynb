Eliminar filas vacÃ­as

```python
# Para que elimine las filas que todos los campos sean null
WorldCupMatches.dropna(how='all', inplace=True)
WorldCupPlayers.dropna(how='all', inplace=True)
WorldCups.dropna(how='all', inplace=True)

```

Eliminando duplicados

```python
duplicados_antes_matches = WorldCupMatches.duplicated(subset='MatchID').sum()
print(f"Total valores duplicas = {duplicados_antes_matches}")
#Boramos los valores duplicados de la tabla matches 
WorldCupMatches.drop_duplicates(subset='MatchID', inplace=True)
```

```python
WorldCupMatches.duplicated().sum()
```

```python

duplicados_antes_players = WorldCupPlayers.duplicated().sum()
print(f"Total valores duplicas = {duplicados_antes_players}")
#Boramos los valores duplicados de la tabla matches 
WorldCupPlayers.drop_duplicates(subset=['MatchID', 'Player Name'],keep='first' ,inplace=True)
```

Estandarizacion de datos

```python
nombres_paises_estandarizados = {
    
    'Germany FR' : 'Germany',
    'German DR' : 'Germany',
    'Soviet Union' : 'Russia',
    'Korea Republic' : 'South Korea',
    'Korea DPR' : 'North Korea ',
    'rn">United Arab Emirates' : 'United Arab Emirates',
    'rn">Republic of Ireland' : 'Republic of Ireland',
    "CÃ¯Â¿Â½te d'Ivoire" : "Ivory Coast",
    'rn">Bosnia and Herzegovina' : 'Bosnia and Herzegovina',
    'rn">Trinidad and Tobago' : 'Trinidad and Tobago',
    'rn">Serbia and Montenegro' : 'Serbia and Montenegro'
    
}

WorldCupMatches['Home Team Name'] = WorldCupMatches['Home Team Name'].replace(nombres_paises_estandarizados)
WorldCupMatches['Away Team Name'] = WorldCupMatches['Away Team Name'].replace(nombres_paises_estandarizados)
WorldCups['Winner'] = WorldCups['Winner'].replace(nombres_paises_estandarizados)
WorldCups['Runners-Up'] = WorldCups['Runners-Up'].replace(nombres_paises_estandarizados)
WorldCups['Third'] = WorldCups['Third'].replace(nombres_paises_estandarizados)
WorldCups['Fourth'] = WorldCups['Fourth'].replace(nombres_paises_estandarizados)
```

```python
Iniciales_paises_estandarizados = {
    'FRG' : 'GER',
    'GDR' : 'GER',
    'URS' : 'RUS',
}

WorldCupPlayers['Team Initials'] = WorldCupPlayers['Team Initials'].replace(Iniciales_paises_estandarizados)
WorldCupMatches['Home Team Initials'] = WorldCupMatches['Home Team Initials'].replace(Iniciales_paises_estandarizados) 
WorldCupMatches['Away Team Initials'] = WorldCupMatches['Away Team Initials'].replace(Iniciales_paises_estandarizados)
```

```python
# Elimiando los puntos que tenian nuestras cantidades de attendance
WorldCups['Attendance'] = WorldCups['Attendance'].str.replace(".","",regex=False).astype(int)
```

```python
nombre_estadios_estandarizados = {
    'Stade VÃ¯Â¿Â½lodrome' : 'Stade VÃ©lodrome',
    'MaracanÃ¯Â¿Â½ - EstÃ¯Â¿Â½dio Jornalista MÃ¯Â¿Â½rio Filho' :'MaracanÃ£ - EstÃ¡dio Jornalista MÃ¡rio Filho' ,
    'Nou Camp - Estadio LeÃ¯Â¿Â½n' : 'Nou Camp - Estadio LeÃ³n',
    'Estadio JosÃ¯Â¿Â½ MarÃ¯Â¿Â½a Minella' : 'Estadio JosÃ© MarÃ­a Minella',
    'Estadio OlÃ¯Â¿Â½mpico Chateau Carreras' : 'Estadio OlÃ­mpico Chateau Carreras',
    'Estadio Municipal de BalaÃ¯Â¿Â½dos' : 'Estadio Municipal de BalaÃ­dos',
    'Estadio OlÃ¯Â¿Â½mpico Universitario' : 'Estadio OlÃ­mpico Universitario',
}

WorldCupMatches['Stadium'] = WorldCupMatches['Stadium'].replace(nombre_estadios_estandarizados)
```

Formateo para SQL

```python

# Quitar espacios vacÃ­os 
WorldCupMatches['Datetime'] = WorldCupMatches['Datetime'].str.strip()

#ConversiÃ³n a Objeto Fecha 

WorldCupMatches['Datetime'] = pd.to_datetime(WorldCupMatches['Datetime'], format='%d %b %Y - %H:%M', errors='coerce')

#  Formateo final para el CSV 
# Una vez que ya es fecha, la volvemos texto con el formato que MySQL 
WorldCupMatches['Datetime'] = WorldCupMatches['Datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')


nulos = WorldCupMatches['Datetime'].isna().sum()
```

```python
nulos
```

```python
# Quita espacios al inicio y final y "saltos de lÃ­nea" que suelen dar problemas en SQL
WorldCupMatches = WorldCupMatches.map(lambda x: x.strip() if isinstance(x, str) else x)
WorldCupPlayers = WorldCupPlayers.map(lambda x: x.strip() if isinstance(x, str) else x)
WorldCups = WorldCups.map((lambda x: x.strip() if isinstance(x, str) else x))
```

```python
#Cambiando el nombre de mis columnas quitando espacion y guiones en medio para que sean sustituidos por guiones bajos
WorldCupMatches.columns = [col.lower().replace(' ', '_').replace('-','_') for col in WorldCupMatches.columns]
WorldCupPlayers.columns = [col.lower().replace(' ', '_').replace('-','_') for col in WorldCupPlayers.columns]
WorldCups.columns = [col.lower().replace(' ', '_').replace('-','_') for col in WorldCups.columns]
```
